```python
from scipy.stats import skew

# Calculate skewness for each column
skewness = data2.apply(skew)

# Categorize based on skewness
highly_right_skewed = skewness[skewness > 1].index    # Strongly right-skewed
moderately_right_skewed = skewness[(skewness > 0.5) & (skewness <= 1)].index  

highly_left_skewed = skewness[skewness < -1].index    # Strongly left-skewed
moderately_left_skewed = skewness[(skewness < -0.5) & (skewness >= -1)].index  
```

```python
import numpy as np


# Apply transformations
data2[highly_right_skewed] = np.log1p(data2[highly_right_skewed])  # Log for right skew
data2[moderately_right_skewed] = np.sqrt(data2[moderately_right_skewed])  # Square root for mild right skew

data2[highly_left_skewed] = np.exp(data2[highly_left_skewed])  # Exponential for left skew
data2[moderately_left_skewed] = np.square(data2[moderately_left_skewed])  # Square for mild left skew
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(25, 35))
for i, column in enumerate(data2.columns, 1):
    plt.subplot(6, 3, i)
    sns.kdeplot(data2[column], color='red')
    plt.title(f'Transformed Distribution of {column}')

plt.tight_layout()
plt.show()
```

```python
sns.heatmap(data2.corr())
```

```python
x = data2.drop(columns=['BALANCE']) 
y = data2['BALANCE'] 
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
y_test
```

```python
lr =  LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r2}")
```
