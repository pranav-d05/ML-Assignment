```python
bins = [10, 20, 30, 40, 50, 60]
labels = ['10-20', '21-30', '31-40', '41-50', '51-60']
age_groups = pd.cut(df['Age'], bins=bins, labels=labels)

grouped = df.groupby(age_groups).agg({
    'Spending Score (1-100)': 'mean',
    'Annual Income (k$)': 'mean'
}).reset_index()

plt.figure(figsize=(8, 5))
ax = sns.barplot( x='Age', y='Spending Score (1-100)', data=grouped, palette='Set2')
for bar, income in zip(ax.patches, grouped['Annual Income (k$)']):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2,height + 1,f"${income:.1f}k",ha='center',fontsize=10,color='black')

plt.title("Average Spending Score by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Spending Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(7, 4))
plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method: Optimal Clusters")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
```

```python
agg = AgglomerativeClustering(n_clusters=5)
agg_labels = agg.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(x=X['Annual Income (k$)'], y=X['Spending Score (1-100)'],
                hue=kmeans_labels, palette='tab10')
plt.title("Clusters by KMeans")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(x=X['Annual Income (k$)'], y=X['Spending Score (1-100)'],
                hue=agg_labels, palette='tab10')
plt.title("Clusters by Agglomerative Clustering")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend(title='Cluster')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import StratifiedKFold

x = np.array([[1,2],[3,4],[1,2],[3,4],[2,3],[2,3]])
y = np.array([0,0,1,1,4,2])
skf = StratifiedKFold(n_splits=2)
for train_index,test_index in skf.split(x,y):
  print("Train:",train_index,"Test:",test_index)
#
```

```python
import numpy as np
from sklearn.model_selection import RepeatedKFold

x = np.array([[1,2],[3,4],[1,2],[3,4],[2,3],[2,3]])
y = np.array([0,0,1,1,4,2])
rkf = RepeatedKFold(n_splits=2,n_repeats=3)
for train_index,test_index in rkf.split(x,y):
  print("Train:",train_index,"Test:",test_index)
```
